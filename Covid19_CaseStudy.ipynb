Q2.2: Generate plots of confirmed cases over time for the top countries.



```python
## We will begin with dropping unecessary columns
## After that we will groupby country and find sum of the confirmed cases in each country.
confirmed_by_country = confirmed.drop(columns=['Province/State', 'Lat', 'Long'])
confirmed_by_country = confirmed_by_country.groupby('Country/Region').sum()

## Transpose the data so dates becomes rows
confirmed_over_time = confirmed_by_country.T

## Top 5 countries for confirmed cases
top_countries = confirmed_by_country.iloc[:, -1].sort_values(ascending=False).head(5).index

## Plotting the data

plt.figure(figsize=(14, 6))

for country in top_countries:
    plt.plot(confirmed_over_time.index, confirmed_over_time[country], label=country)

plt.xlabel('Date', fontsize=12)
plt.ylabel('Confirmed Cases', fontsize=12)
plt.title('COVID-19 Confirmed Cases Over Time (Top 5 Countries)', fontsize=14)

# Format x-axis ticks: show every 30th date
plt.xticks(confirmed_over_time.index[::30], rotation=45)

# Legend outside the plot
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title_fontsize=11)

plt.tight_layout()
plt.grid(True)
plt.show()
```

Q2.3: Generate plots of confirmed cases over time for China.

```python
## We will use line chart for plotting confirmed cases in china

china_df = confirmed[confirmed['Country/Region'] == 'China']

# Drop non-date columns
non_date_cols = ['Province/State', 'Country/Region', 'Lat', 'Long']
china_df = china_df.drop(columns=non_date_cols)

# Now all remaining columns should be date strings like '1/22/20'
# Sum all provinces to get China's total cases per day
china_total = china_df.sum()

# Safely convert the column names (dates) to datetime
china_total.index = pd.to_datetime(china_total.index, errors='coerce')

# Drop any invalid/missing datetime conversions (if any)
china_total = china_total.dropna()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(china_total.index, china_total.values, color='red', label='China')

plt.title('COVID-19 Confirmed Cases Over Time in China', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Confirmed Cases', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

Q5.1: Analyze the peak number of daily new cases in Germany, France, and Italy. Which country experienced the highest single-day surge, and when did it occur?



```python
## Melt the dataframe to transform data of columns into rows

confirmed_melted = confirmed.melt(
    id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name = 'Date',
    value_name = 'Confirmed'
)

## Convert date column into datetime object

confirmed_melted['Date'] = pd.to_datetime(confirmed_melted['Date'])

## Filter data in Germany,France, and Italy

countries = ['Germany' ,'France', 'Italy']
filtered_data = confirmed_melted[confirmed_melted['Country/Region'].isin(countries)].copy()

## Sum up all confirmed cases for a country on a given date, regardless of province/state

country_total_confirmed = filtered_data.groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index()
country_total_confirmed.sort_values(by=['Country/Region', 'Date'], inplace=True)

# Calculate daily new cases for each country
# The .diff() method calculates the difference between the current and a prior element.
# The .fillna(0) is used to handle the first day for each country, where the difference would be NaN.

country_total_confirmed['Daily_new_cases'] = country_total_confirmed.groupby('Country/Region')['Confirmed'].diff().fillna(0)

# Convert Daily_New_Cases to integer type and ensure no negative values
country_total_confirmed['Daily_new_cases'] = country_total_confirmed['Daily_new_cases'].apply(lambda x: max(0, int(x)))

## Peak daily new cases for each country
peak_daily_surge = country_total_confirmed.loc[country_total_confirmed.groupby('Country/Region')['Daily_new_cases'].idxmax()]

print("Peak Daily New Cases for France, Germany, Italy: ")
print(peak_daily_surge)

highest_surge = peak_daily_surge.loc[peak_daily_surge['Daily_new_cases'].idxmax()]

print("\n Country with highest daily surge: ")
print(highest_surge)
```

Q5.2: Compare the recovery rates (recoveries/confirmed cases) between Canada and Australia as of December 31, 2020. Which country showed better management of the pandemic according to this metric?

```python
# Column names for all dataframes to identify the correct column names
confirmed.columns.tolist()
deaths.columns.tolist()
recovered.columns.tolist()

# Define the date for analysis
analysis_date = '12/31/20'

# Filter data for Canada and Australia for confirmed cases
canada_confirmed = confirmed[confirmed['Country/Region'] == 'Canada']
australia_confirmed = confirmed[confirmed['Country/Region'] == 'Australia']

# Filter data for Canada and Australia for recovered cases
# Assuming 'Country/Region' and the date column are correctly identified after re-loading.
# If the previous print statements show 'ColumnX' as country, this will still fail.
canada_recovered = recovered[recovered['Country/Region'] == 'Canada']
australia_recovered = recovered[recovered['Country/Region'] == 'Australia']

# Get the total confirmed cases for Canada and Australia as of December 31, 2020
total_confirmed_canada = canada_confirmed[analysis_date].sum()
total_confirmed_australia = australia_confirmed[analysis_date].sum()

# Get the total recovered cases for Canada and Australia as of December 31, 2020
total_recovered_canada = canada_recovered[analysis_date].sum()
total_recovered_australia = australia_recovered[analysis_date].sum()

# Calculate the recovery rate for Canada
# Avoid division by zero
if total_confirmed_canada > 0:
    recovery_rate_canada = (total_recovered_canada / total_confirmed_canada) * 100
else:
    recovery_rate_canada = 0

# Calculate the recovery rate for Australia
# Avoid division by zero
if total_confirmed_australia > 0:
    recovery_rate_australia = (total_recovered_australia / total_confirmed_australia) * 100
else:
    recovery_rate_australia = 0

print(f"Confirmed cases in Canada as of {analysis_date}: {total_confirmed_canada}")
print(f"Recovered cases in Canada as of {analysis_date}: {total_recovered_canada}")
print(f"Recovery rate in Canada as of {analysis_date}: {recovery_rate_canada:.2f}%")

print(f"\nConfirmed cases in Australia as of {analysis_date}: {total_confirmed_australia}")
print(f"Recovered cases in Australia as of {analysis_date}: {total_recovered_australia}")
print(f"Recovery rate in Australia as of {analysis_date}: {recovery_rate_australia:.2f}%")

if recovery_rate_canada > recovery_rate_australia:
    better_management = "Canada"
elif recovery_rate_australia > recovery_rate_canada:
    better_management = "Australia"
else:
    better_management = "Both countries had similar management"

print(f"\nAccording to the recovery rate metric, {better_management} showed better management of the pandemic as of {analysis_date}.")
```

Q5.3: What is the distribution of death rates (deaths/confirmed cases) among provinces in Canada? Identify the province with the highest and lowest death rate as of the latest data point.

```python
 latest_date_confirmed = confirmed.columns[-1]

 latest_date_deaths = deaths.columns[-1]

 canada_confirmed_data = confirmed[confirmed['Country/Region']== 'Canada']
 canada_deaths_data = deaths[deaths['Country/Region']=='Canada']

 province_confirmed = canada_confirmed_data.groupby('Province/State')[latest_date_confirmed].sum()

 province_deaths = canada_deaths_data.groupby('Province/State')[latest_date_deaths].sum()

 death_rates_canada = pd.DataFrame({
     'Confirmed' : province_confirmed,
     'Deaths': province_deaths
})

# Calculate the death rate, handling division by zero
death_rates_canada['Death Rate (%)'] = death_rates_canada.apply(
    lambda row: (row['Deaths'] / row['Confirmed']) * 100 if row['Confirmed'] > 0 else 0,
    axis=1
)

# Sort by death rate for distribution analysis
death_rates_canada_sorted = death_rates_canada.sort_values(by='Death Rate (%)', ascending=False)

# Identify the province with the highest death rate
highest_death_rate_province = death_rates_canada_sorted.iloc[0]
highest_province_name = highest_death_rate_province.name
highest_death_rate_value = highest_death_rate_province['Death Rate (%)']

# Identify the province with the lowest death rate (excluding those with 0 confirmed cases if applicable)
lowest_death_rate_province = death_rates_canada_sorted[death_rates_canada_sorted['Confirmed'] > 0].iloc[-1]
lowest_province_name = lowest_death_rate_province.name
lowest_death_rate_value = lowest_death_rate_province['Death Rate (%)']

print(f"Distribution of Death Rates Among Provinces in Canada (as of {latest_date_confirmed}):")
print(death_rates_canada_sorted[['Deaths', 'Confirmed', 'Death Rate (%)']].to_string())

print(f"\nProvince with the Highest Death Rate:")
print(f"Province: {highest_province_name}")
print(f"Death Rate: {highest_death_rate_value:.2f}%")
print(f"Deaths: {highest_death_rate_province['Deaths']}")
print(f"Confirmed Cases: {highest_death_rate_province['Confirmed']}")

print(f"\nProvince with the Lowest Death Rate:")
print(f"Province: {lowest_province_name}")
print(f"Death Rate: {lowest_death_rate_value:.2f}%")
print(f"Deaths: {lowest_death_rate_province['Deaths']}")
print(f"Confirmed Cases: {lowest_death_rate_province['Confirmed']}")
```

Question 6: Data Transformation

Q6.1: Transform the 'deaths' dataset from wide format (where each column represents a date) to long format, where each row represents a single date, ensuring that the date column is in datetime format. How would this transformation be executed?

```python
id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long']

date_columns_deaths = deaths.columns.difference(id_vars)

deaths_long = deaths.melt(
    id_vars = id_vars,
    value_vars = date_columns_deaths,
    var_name = 'Date',
    value_name = 'Deaths'
)

deaths_long['Date'] = pd.to_datetime(deaths_long['Date'], errors='coerce')

print("Transformed Deaths Data (long format - Head)")
print(deaths_long.head())

print("Transformed Deaths (long format - Info)")
print(deaths_long.info())
```

Q6.2: What is the total number of deaths reported per country up to the current date?



```python
latest_date = deaths_long['Date'].max()

deaths_till_latest_date = deaths_long[deaths_long['Date'] <= latest_date]

total_deaths_per_country = deaths_till_latest_date.groupby('Country/Region')['Deaths'].sum().reset_index()

deaths_per_country_sorted = total_deaths_per_country.sort_values(by='Deaths', ascending=False)

print(f"Total number of deaths reported upto: {latest_date.strftime('%Y-%m-%d')}")

print(deaths_per_country_sorted.to_string(index=False))
```

Q6.3: What are the top 5 countries with the highest average daily deaths?



```python

correct_columns = confirmed.columns.tolist()
# Assign the correct column names to the deaths DataFrame
deaths.columns = correct_columns

# Remove the first row which is now a redundant header
df_deaths = deaths.iloc[1:].copy()

# Ensure numeric conversion for date columns
date_columns = correct_columns[4:]
for col in date_columns:
    df_deaths[col] = pd.to_numeric(df_deaths[col], errors='coerce').fillna(0)


# Melt the dataframe from wide to long format
id_vars_list = ['Province/State', 'Country/Region', 'Lat', 'Long']
df_deaths_long = df_deaths.melt(id_vars=id_vars_list, var_name='Date', value_name='Deaths')

# Convert the 'Date' column to datetime objects
df_deaths_long['Date'] = pd.to_datetime(df_deaths_long['Date'], format='%m/%d/%y')

# Convert 'Deaths' column to numeric
df_deaths_long['Deaths'] = pd.to_numeric(df_deaths_long['Deaths'], errors='coerce')

# Group by country and date, summing the deaths to get a single value per country per day
df_deaths_daily = df_deaths_long.groupby(['Country/Region', 'Date'])['Deaths'].sum().reset_index()

# Calculate the daily deaths (difference from the previous day)
df_deaths_daily['Daily Deaths'] = df_deaths_daily.groupby('Country/Region')['Deaths'].diff().fillna(0)

# Calculate the average daily deaths for each country
average_daily_deaths = df_deaths_daily.groupby('Country/Region')['Daily Deaths'].mean()

# Get the top 5 countries with the highest average daily deaths
top_5_countries_daily_deaths = average_daily_deaths.nlargest(5)

# Display the result
print("Top 5 countries with the highest average daily deaths:")
print(top_5_countries_daily_deaths.to_markdown(numalign="left", stralign="left"))
```

Q6.4: How have the total deaths evolved over time in the United States?

```python
correct_columns = confirmed.columns.tolist()

# Assign the correct column names to the deaths DataFrame
deaths.columns = correct_columns

# Remove the first row which is now a redundant header
df_deaths = deaths.iloc[1:].copy()

# Ensure numeric conversion for date columns
date_columns = correct_columns[4:]
for col in date_columns:
    df_deaths[col] = pd.to_numeric(df_deaths[col], errors='coerce').fillna(0)


# Melt the dataframe from wide to long format
id_vars_list = ['Province/State', 'Country/Region', 'Lat', 'Long']
df_deaths_long = df_deaths.melt(id_vars=id_vars_list, var_name='Date', value_name='Deaths')

# Convert the 'Date' column to datetime objects
df_deaths_long['Date'] = pd.to_datetime(df_deaths_long['Date'], format='%m/%d/%y')

# Filter for the United States
df_us_deaths = df_deaths_long[df_deaths_long['Country/Region'] == 'US'].copy()

# Group by date and sum the deaths to get total deaths for the US each day
df_us_total_deaths = df_us_deaths.groupby('Date')['Deaths'].sum().reset_index()

# Display the first 5 rows of the processed data
print("First 5 rows of US total deaths over time:")
print(df_us_total_deaths.head().to_markdown(index=False, numalign="left", stralign="left"))
```
